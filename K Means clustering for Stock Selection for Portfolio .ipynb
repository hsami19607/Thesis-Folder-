```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('C:\\Users\\Owner\Desktop\Ratiosecond.csv')
df
```

```python
plt.scatter(df['ast_turn'], df['oprt_mrgn'])
plt.xlabel('Asset Turnover Ratio')
plt.ylabel('Operating Margin Ratio')
plt.show()
```

```python
x = df.copy()
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['cluster_pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['ast_turn'], clusters['oprt_mrgn'], c = clusters['cluster_pred'], cmap = 'rainbow')
plt.xlabel('Asset Turnover Ratio')
plt.ylabel('Operating Margin Ratio')
plt.show()
```

```python
from sklearn import preprocessing 
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
weights = []

for i in range(1, 30):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    weights.append(kmeans.inertia_)
weights
```

```python
plt.plot(range(1, 30), weights)
plt.xlabel('Number of Clsuters')
plt.ylabel('Weights')
plt.show()
```

```python
kmeans_new = KMeans(4)
kmeans.fit(x_scaled)
cluster_new = x.copy()
cluster_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
cluster_new
```

```python
plt.scatter(cluster_new['ast_turn'], cluster_new['oprt_mrgn'], c = cluster_new['cluster_pred'], cmap = 'rainbow')
plt.xlabel('Asset Turnover Ratio')
plt.ylabel('Oper')
```
